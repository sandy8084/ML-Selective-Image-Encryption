```python
import pandas as pd
import os
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
```

```python
#-------------------KNN MODEL-----------------------------#
# -----------------------------
# Base paths
# -----------------------------
BASE_DIR = ".."
data_dir = os.path.join(BASE_DIR, "data")
models_dir = os.path.join(BASE_DIR, "models")
results_dir = os.path.join(BASE_DIR, "results")

os.makedirs(models_dir, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)

# -----------------------------
# Load training data
# -----------------------------
X_train = pd.read_csv(os.path.join(data_dir, "X_train.csv"))
y_train = pd.read_csv(os.path.join(data_dir, "y_train.csv")).values.ravel()

# -----------------------------
# Load testing data
# -----------------------------
X_test = pd.read_csv(os.path.join(data_dir, "X_test.csv"))
y_test = pd.read_csv(os.path.join(data_dir, "y_test.csv")).values.ravel()

# -----------------------------
# Train KNN model
# -----------------------------
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# -----------------------------
# Save model
# -----------------------------
model_path = os.path.join(models_dir, "knn_trained_model.pkl")
joblib.dump(knn, model_path)

print(f"KNN model saved at {model_path}")

# -----------------------------
# Predictions
# -----------------------------
y_pred = knn.predict(X_test)

# -----------------------------
# Evaluation
# -----------------------------
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"KNN Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", cm)

# -----------------------------
# Save confusion matrix image
# -----------------------------
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Encrypt', 'Encrypt'],
            yticklabels=['No Encrypt', 'Encrypt'])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("KNN Confusion Matrix")

cm_path = os.path.join(results_dir, "knn_confusion_matrix.png")
plt.savefig(cm_path)
plt.close()

print(f"KNN confusion matrix saved at {cm_path}")
# -----------------------------
# Overfitting check
# -----------------------------
train_accuracy = knn.score(X_train, y_train) * 100
test_accuracy =  knn.score(X_test, y_test) * 100

print(f"\nTraining Accuracy: {train_accuracy:.2f}%")
print(f"Testing Accuracy: {test_accuracy:.2f}%")

if train_accuracy - test_accuracy > 5:
    print("Possible Overfitting Detected. Consider regularization or more data.")
else:
    print("No major signs of overfitting.")
```

```python
#---------------------- Naives Bayes --------------------------#

# -----------------------------
# Base paths
# -----------------------------
BASE_DIR = ".."
data_dir = os.path.join(BASE_DIR, "data")
models_dir = os.path.join(BASE_DIR, "models")
results_dir = os.path.join(BASE_DIR, "results")

os.makedirs(models_dir, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)

# -----------------------------
# Load training data
# -----------------------------
X_train = pd.read_csv(os.path.join(data_dir, "X_train.csv"))
y_train = pd.read_csv(os.path.join(data_dir, "y_train.csv")).values.ravel()

# -----------------------------
# Load testing data
# -----------------------------
X_test = pd.read_csv(os.path.join(data_dir, "X_test.csv"))
y_test = pd.read_csv(os.path.join(data_dir, "y_test.csv")).values.ravel()

# -----------------------------
# Train Naive Bayes model
# -----------------------------
nb = GaussianNB()
nb.fit(X_train, y_train)

# -----------------------------
# Save model
# -----------------------------
model_path = os.path.join(models_dir, "naivebayes_trained_model.pkl")
joblib.dump(nb, model_path)

print(f"Naive Bayes model saved at {model_path}")

# -----------------------------
# Predictions
# -----------------------------
y_pred = nb.predict(X_test)

# -----------------------------
# Evaluation
# -----------------------------
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Naive Bayes Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", cm)

# -----------------------------
# Save confusion matrix image
# -----------------------------
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Encrypt', 'Encrypt'],
            yticklabels=['No Encrypt', 'Encrypt'])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Naive Bayes Confusion Matrix")

cm_path = os.path.join(results_dir, "naivebayes_confusion_matrix.png")
plt.savefig(cm_path)
plt.close()

print(f"Naive Bayes confusion matrix saved at {cm_path}")
# -----------------------------
# Overfitting check
# -----------------------------
train_accuracy = nb.score(X_train, y_train) * 100
test_accuracy = nb.score(X_test, y_test) * 100

print(f"\nTraining Accuracy: {train_accuracy:.2f}%")
print(f"Testing Accuracy: {test_accuracy:.2f}%")

if train_accuracy - test_accuracy > 5:
    print("Possible Overfitting Detected. Consider regularization or more data.")
else:
    print("No major signs of overfitting.")
```

```python
#----------- ROC CURVE FOR THREE MODELS ---------------------#
# -----------------------------
# Base paths
# -----------------------------
BASE_DIR = ".."
data_dir = os.path.join(BASE_DIR, "data")
models_dir = os.path.join(BASE_DIR, "models")
results_dir = os.path.join(BASE_DIR, "results")

os.makedirs(results_dir, exist_ok=True)

# -----------------------------
# Load test data
# -----------------------------
X_test = pd.read_csv(os.path.join(data_dir, "X_test.csv"))
y_test = pd.read_csv(os.path.join(data_dir, "y_test.csv")).values.ravel()

# Convert to binary (needed for ROC)
y_test_bin = label_binarize(y_test, classes=[0, 1]).ravel()

# -----------------------------
# Load models
# -----------------------------
models = {
    "SVM": joblib.load(os.path.join(models_dir, "svm_trained_model.pkl")),
    "KNN": joblib.load(os.path.join(models_dir, "knn_trained_model.pkl")),
    "Naive Bayes": joblib.load(os.path.join(models_dir, "naivebayes_trained_model.pkl")),
}

# -----------------------------
# Plot ROC curves
# -----------------------------
plt.figure(figsize=(8, 6))

for model_name, model in models.items():

    if hasattr(model, "predict_proba"):
        y_probs = model.predict_proba(X_test)[:, 1]
    else:
        y_probs = model.decision_function(X_test)

    fpr, tpr, _ = roc_curve(y_test_bin, y_probs)
    roc_auc = auc(fpr, tpr)

    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {roc_auc:.2f})")

# Random baseline
plt.plot([0, 1], [0, 1], 'k--', label="Random Guess (AUC = 0.50)")

# Labels
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve Comparison")
plt.legend()
plt.grid()

# -----------------------------
# Save plot (IMPORTANT)
# -----------------------------
roc_path = os.path.join(results_dir, "roc_curve.png")
plt.savefig(roc_path)
plt.close()

print(f"ROC curve saved at {roc_path}")
```
