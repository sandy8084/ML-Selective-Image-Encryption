```python
import pandas as pd
import os
import joblib
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
```

```python
#-------------------KNN MODEL-----------------------------#
# -----------------------------
# Base paths
# -----------------------------
BASE_DIR = ".."
data_dir = os.path.join(BASE_DIR, "data")
models_dir = os.path.join(BASE_DIR, "models")
results_dir = os.path.join(BASE_DIR, "results")

os.makedirs(models_dir, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)

# -----------------------------
# Load training data
# -----------------------------
X_train = pd.read_csv(os.path.join(data_dir, "X_train.csv"))
y_train = pd.read_csv(os.path.join(data_dir, "y_train.csv")).values.ravel()

# -----------------------------
# Load testing data
# -----------------------------
X_test = pd.read_csv(os.path.join(data_dir, "X_test.csv"))
y_test = pd.read_csv(os.path.join(data_dir, "y_test.csv")).values.ravel()

# -----------------------------
# Train KNN model
# -----------------------------
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# -----------------------------
# Save model
# -----------------------------
model_path = os.path.join(models_dir, "knn_trained_model.pkl")
joblib.dump(knn, model_path)

print(f"KNN model saved at {model_path}")

# -----------------------------
# Predictions
# -----------------------------
y_pred = knn.predict(X_test)

# -----------------------------
# Evaluation
# -----------------------------
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"KNN Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", cm)

# -----------------------------
# Save confusion matrix image
# -----------------------------
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Encrypt', 'Encrypt'],
            yticklabels=['No Encrypt', 'Encrypt'])

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("KNN Confusion Matrix")

cm_path = os.path.join(results_dir, "knn_confusion_matrix.png")
plt.savefig(cm_path)
plt.close()

print(f"KNN confusion matrix saved at {cm_path}")
```

```python
# Load training data
X_train = pd.read_csv("X_train.csv")
y_train = pd.read_csv("y_train.csv").values.ravel()  # Convert to 1D array

# Load testing data
X_test = pd.read_csv("X_test.csv")
y_test = pd.read_csv("y_test.csv").values.ravel()  # Convert to 1D array

# Train Naïve Bayes model
nb = GaussianNB()
nb.fit(X_train, y_train)

# Save the trained model
model_filename = "naivebayes_trained_model.pkl"
joblib.dump(nb, model_filename)
print(f" Model trained and saved as '{model_filename}'")

# Predict labels
y_pred = nb.predict(X_test)
print(f" Naïve Bayes Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")
print("\n Classification Report:\n", classification_report(y_test, y_pred))
print("\n Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# Get training accuracy
train_accuracy = nb.score(X_train, y_train) * 100
test_accuracy = nb.score(X_test, y_test) * 100

print(f"🔹 Training Accuracy: {train_accuracy:.2f}%")
print(f"🔹 Testing Accuracy: {test_accuracy:.2f}%")

# Check if there's a large gap
if train_accuracy - test_accuracy > 5:
    print(" Possible Overfitting Detected! Consider regularization or more data.")
else:
    print(" No major signs of overfitting.")
```

```python
#ROC CURVES
# Load dataset
X_test = pd.read_csv("X_test.csv")
y_test = pd.read_csv("y_test.csv").values.ravel()  # Convert to 1D array

# Convert labels to binary format (needed for ROC curve)
y_test_bin = label_binarize(y_test, classes=[0, 1]).ravel()

# Load trained models
models = {
    "SVM": joblib.load("svm_trained_model.pkl"),
    "KNN": joblib.load("knn_trained_model.pkl"),
    "Naïve Bayes": joblib.load("naivebayes_trained_model.pkl"),
}

# Plot ROC Curve for each model
plt.figure(figsize=(8, 6))

for model_name, model in models.items():
    # Get probability scores
    if hasattr(model, "predict_proba"):
        y_probs = model.predict_proba(X_test)[:, 1]  # Probability of class 1
    else:
        y_probs = model.decision_function(X_test)  # For SVM decision function

    # Compute FPR, TPR, and AUC
    fpr, tpr, _ = roc_curve(y_test_bin, y_probs)
    roc_auc = auc(fpr, tpr)

    # Plot ROC Curve
    plt.plot(fpr, tpr, label=f"{model_name} (AUC = {roc_auc:.2f})")

# Plot random chance line
plt.plot([0, 1], [0, 1], 'k--', label="Random Guess (AUC = 0.50)")

# Customize plot
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("ROC Curve Comparison Using Trained Models")
plt.legend()
plt.grid()
plt.show()
```
