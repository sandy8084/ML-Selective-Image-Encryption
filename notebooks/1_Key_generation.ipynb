```python
import os

a, b = 1.4, 0.3
iterations = 65536  
x, y = 0.1, 0.3

# -----------------------------
# Generate Henon Map
# -----------------------------
x_values, y_values = [], []

for _ in range(iterations):
    x_next = 1 - a * x**2 + y
    y_next = b * x
    x, y = x_next, y_next
    x_values.append(x)
    y_values.append(y)

# -----------------------------
# Normalize to 0–255
# -----------------------------
x_norm = [int(((val - min(x_values)) / (max(x_values) - min(x_values))) * 255) for val in x_values]
y_norm = [int(((val - min(y_values)) / (max(y_values) - min(y_values))) * 255) for val in y_values]

# -----------------------------
# Save to keys/ folder
# -----------------------------
BASE_DIR = ".."   # because notebook is inside notebooks/
keys_dir = os.path.join(BASE_DIR, "keys")

os.makedirs(keys_dir, exist_ok=True)

x_path = os.path.join(keys_dir, "henon_x.txt")
y_path = os.path.join(keys_dir, "henon_y.txt")

with open(x_path, "w") as fx:
    fx.write("\n".join(map(str, x_norm)))

with open(y_path, "w") as fy:
    fy.write("\n".join(map(str, y_norm)))

print(f"Henon keys saved to:\n{x_path}\n{y_path}")
```
