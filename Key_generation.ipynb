```python
a, b = 1.4, 0.3  # Standard Henon map constants
iterations = 65536  
x, y = 0.1, 0.3  # Initial conditions (seed)

# Generate 2D Henon Map values
x_values, y_values = [], []
for _ in range(iterations):
    x_next = 1 - a * x**2 + y
    y_next = b * x
    x, y = x_next, y_next
    x_values.append(x)
    y_values.append(y)

# Normalize values to range 0-255
x_norm = [int(((val - min(x_values)) / (max(x_values) - min(x_values))) * 255) for val in x_values]
y_norm = [int(((val - min(y_values)) / (max(y_values) - min(y_values))) * 255) for val in y_values]

# Save x values in a continuous column (one value per line)
with open("henon_x.txt", "w") as fx:
    fx.write("\n".join(map(str, x_norm)))  # Newline-separated values

# Save y values in a continuous column (one value per line)
with open("henon_y.txt", "w") as fy:
    fy.write("\n".join(map(str, y_norm)))  # Newline-separated values

print("Henon map values saved in a continuous column format to 'henon_x.txt' and 'henon_y.txt'.")
```
